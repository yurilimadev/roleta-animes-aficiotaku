# TESTE - Extração

```python
# Supondo que a div pai tenha um ID exclusivo, ajuste conforme necessário
div_pai_xpath = '//*[@id="body_wrapper"]/div[5]/div/div[1]/div[1]/div[1]/ol'

# Adapte a função extrair_informacoes com o XPath da div pai
div_pai = driver.find_element(By.XPATH, div_pai_xpath)

def extrair_informacoes(div_pai):
    elementos_processados = set()
    informacoes = []

    while True:
        # Encontrar todos os elementos li dentro da div pai
        elementos_li = div_pai.find_elements(By.CLASS_NAME, 'span2.movie_list_item')

        for elemento_li in elementos_li:
            elemento_id = elemento_li.get_attribute('id')
            if elemento_id not in elementos_processados:
                nome = elemento_li.find_element(By.CLASS_NAME, 'title').text
                src = elemento_li.find_element(By.TAG_NAME, 'img').get_attribute('data-src')
                informacoes.append({'nome': nome, 'src': src})
                print(nome, src)
                elementos_processados.add(elemento_id)

        # Obter a quantidade inicial de elementos na div pai
        quantidade_elementos_inicial = len(elementos_li)
        print(quantidade_elementos_inicial)

        # Tentar clicar no botão "Carregar mais"
        try:
            botao_carregar_mais = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="body_wrapper"]/div[5]/div/div[1]/div[1]/div[1]/div/button'))
            )
            driver.execute_script("arguments[0].scrollIntoView();", botao_carregar_mais)
            botao_carregar_mais.click()
        except TimeoutException:
            break  # Se o botão "Carregar mais" não estiver presente, encerra o loop

        # Aguardar até que novos elementos sejam adicionados à div pai
        WebDriverWait(driver, 10).until(
            lambda driver: len(driver.find_elements(By.CLASS_NAME, 'span2.movie_list_item')) > quantidade_elementos_inicial
        )
        time.sleep(2)  # Aguarda um tempo adicional (ajuste conforme necessário)

    return informacoes
```

```python
informacoes_imagens = extrair_informacoes(div_pai)
```
